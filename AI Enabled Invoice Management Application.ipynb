## Milestone 1--Data Preprocessing

```python
import numpy as np # linear algebra
import pandas as pd # data processing
import datetime as dt
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import VarianceThreshold as vt
from sklearn.preprocessing import LabelEncoder
from mlxtend.feature_selection import SequentialFeatureSelector
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import seaborn as sns
import matplotlib.pyplot as plt
import collections

#Training data Import
train_data=pd.read_csv('H2HBABBA3057.csv')
train_data.head()
#train_data.dropna(axis=0)
```

## CLEANING PROCESS

```python
#Removing Constant Columns ->> Columns which have the same value in the enitre dataset
unique_cols =  [x for x in train_data.columns if train_data[x].nunique()==1] 
print(unique_cols)
train_data.drop(unique_cols,axis=1,inplace=True)

#renaming columns
train_data.rename(columns={'buisness_year':'business_year'}, inplace=True)
train_data.rename(columns={'document_create_date.1':'document_create_date_1'}, inplace=True)

#Converting datatypes
train_data['document_create_date']=train_data.document_create_date.astype(object)
train_data['document_create_date_1']=train_data.document_create_date_1.astype(object)
train_data['baseline_create_date']=train_data.baseline_create_date.astype(int)
train_data['business_year']=train_data.business_year.astype(object)
#train_data['business_code']=train_data.business_code.astype('category')



#Converting data type to datetime
train_data['clear_date']=pd.to_datetime(train_data.clear_date)
train_data['posting_date']=pd.to_datetime(train_data.posting_date)



#removing columns
train_data.drop('area_business',axis=1,inplace=True)

#Date Conversions for columns
train_data['document_create_date_1']=train_data.document_create_date.astype(object)
document_create_date_1_date_conversion = list(train_data.document_create_date_1)
for i in range(len(document_create_date_1_date_conversion)):
    date = str(document_create_date_1_date_conversion[i])
    str_date = date[0:4]+'-'+date[4:6]+'-'+date[6:8]
    train_data.at[i,'document_create_date_1']=str_date

train_data['document_create_date']=train_data.document_create_date.astype(object)
document_create_date_date_conversion = list(train_data.document_create_date)
for i in range(len(document_create_date_date_conversion)):
    date = str(document_create_date_date_conversion[i])
    str_date = date[0:4]+'-'+date[4:6]+'-'+date[6:8]
    train_data.at[i,'document_create_date']=str_date
    
train_data['due_in_date']=train_data.due_in_date.astype(object)
due_in_date_date_conversion = list(train_data.due_in_date)
for i in range(len(due_in_date_date_conversion)):
    date = str(due_in_date_date_conversion[i])
    str_date = date[0:4]+'-'+date[4:6]+'-'+date[6:8]
    train_data.at[i,'due_in_date']=str_date
    
train_data['baseline_create_date']=train_data.baseline_create_date.astype(object)
baseline_create_date_date_conversion = list(train_data.baseline_create_date)
for i in range(len(baseline_create_date_date_conversion)):
    date = str(baseline_create_date_date_conversion[i])
    str_date = date[0:4]+'-'+date[4:6]+'-'+date[6:8]
    train_data.at[i,'baseline_create_date']=str_date
    
train_data['business_year']=train_data.business_year.astype(object)
business_year_date_conversion = list(train_data.business_year)
for i in range(len(business_year_date_conversion)):
    date = str(business_year_date_conversion[i])
    str_date = date[0:4]
    train_data.at[i,'business_year']=str_date

train_data['due_in_date']=pd.to_datetime(train_data.due_in_date)
```

# DATA PREPROCESSING


```python
#Sorting based on document_create_date_1 column
train_data.sort_values(['document_create_date_1'],ascending=True,inplace=True)
train_data.reset_index(inplace=True, drop=True)
train_data
```

```python
#removing Row which contain Invoice_id as invalid or which has NaN as value
row_remover= train_data[train_data['invoice_id'].isnull()].index.tolist()
train_data.drop(row_remover,axis=0, inplace=True)

#removing any duplicate rows from the dataset
train_data[train_data.duplicated()]
train_data.drop_duplicates(keep='first',inplace=True)

train_data.isnull().sum()
```
